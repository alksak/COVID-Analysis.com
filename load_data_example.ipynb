```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Load Data for specific Date

```python
country = 'Chile'
projection = pd.read_json('projections/'+country+'/'+country+'_pred_26_5_2020.json')
```

## Dataframe

```python
projection
```

## Cummulative Projected Cases

```python
plt.rcParams['figure.figsize'] = [14, 6]
plt.figure()

plt.plot(projection['Total_cases_pred'])
```

## Daily Projected Cases

```python
plt.rcParams['figure.figsize'] = [14, 6]
plt.figure()

plt.plot(projection['Daily_cases_pred'])
```
